# Dataset 1

```python
import pandas as pd
file_path = '/content/drive/MyDrive/Data Mining/Table1.txt'
df = pd.read_csv(file_path, delimiter = ' ', header=None)
df.isna()
df=df.dropna(axis=1)
df
```

## Principle Component Analysis

```python
pip install pca
```

```python
# Calculate the mean vector for the dataset
mean_vector1 = df.mean(axis=0)

# Create a difference matrix by subtracting the mean vector from each row in the normalized dataset
difference_matrix1 = df - mean_vector1

# Compute the covariance matrix C by multiplying the difference matrix D by its transpose D.T
covariance_matrix1 = difference_matrix1.T.dot(difference_matrix1)
covariance_matrix1
```

```python
import numpy as np
covariance_matrix1 = covariance_matrix1.drop(covariance_matrix1.index[-1])

# Drop the last column
covariance_matrix1 = covariance_matrix1.drop(covariance_matrix1.columns[-1], axis=1)
# covariance_matrix =
matrix1= covariance_matrix1.to_numpy()

# Compute eigenvalues and eigenvectors
eigenvalues1, eigenvectors1 = np.linalg.eig(matrix1)
eigenvalues1
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(eigenvalues1) + 1), eigenvalues1, color='blue')
plt.title('Eigenvalues')
plt.xlabel('Index', fontsize=12)
plt.ylabel('Eigenvalue', fontsize=12)
plt.xticks(np.arange(1, len(eigenvalues1) + 1, 1), fontsize=5)
plt.yticks(fontsize=10)
plt.grid(True)
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=10)

df=df.dropna(axis=1)
red_Data = pca.fit_transform(df)
red_Data.shape
```

```python
plt.figure(figsize=(8, 4))
plt.imshow(df)
plt.colorbar(label='Value')
plt.title('Visualization of Dataset 1')
plt.xlabel('Dimensions')
plt.ylabel('N')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.imshow(red_Data)
plt.colorbar(label='Value')
plt.title('Visualization of Reduced Data by PCA')
plt.xlabel('Dimensions')
plt.ylabel('Principal Component Index')
plt.show()
```

```python
# Reconstruct the original data from its reduced representation
reconstructed_data = pca.inverse_transform(red_Data)

plt.figure(figsize=(8, 4))
plt.imshow(reconstructed_data)
plt.colorbar(label='Value')
plt.title('Visualization of Reconstructed Image')
plt.xlabel('Dimensions')
plt.ylabel('Principal Component Index')
plt.show()
```

## Discrete Cosine Transform

```python
import numpy as np
import pandas as pd
from scipy.fftpack import dct, idct
import matplotlib.pyplot as plt
data1_array = df.to_numpy()

dct_result =dct(dct(data1_array, axis=0, norm='ortho'), axis=1, norm='ortho')

threshold1 = np.percentile(np.abs(dct_result), 95)

dct_reduced1 = dct_result * (np.abs(dct_result) > threshold1)
idct_result=  idct(idct(dct_reduced1, axis=0, norm='ortho'), axis=1, norm='ortho')

reconstructed_image1 = idct_result
reconstructed_image1.shape
```

```python
plt.figure(figsize=(15, 10))

plt.subplot(3, 2, 1)
plt.imshow(df)
plt.title('Original Image')
plt.axis('off')

plt.subplot(3, 2, 2)
plt.imshow(reconstructed_image1)
plt.title('Image with reduced dimensions')
plt.axis('off')

plt.show()
```

```python
#DCT by plotting the rows
dct_result=dct(df, norm="ortho", axis=1)
df_dct = pd.DataFrame(dct_result)
df_dct
```

```python
plt.figure(figsize=(10, 6))
for index, row in df_dct.iterrows():
    plt.plot(row, label=f'Row {index}')

plt.xlabel('DCT Coefficient Index')
plt.ylabel('DCT Coefficient Value')
plt.title('DCT Coefficients Line Plot')

plt.grid(True)
plt.show()
```

```python
n_components = 10  # Number of components you wish to keep

# Create a copy to modify
dct_mod = np.copy(dct_result)

# Zero out all but the 'n_components' largest coefficients in each row
for i in range(dct_mod.shape[0]):
    # Find the 'n_components'th largest coefficient value
    threshold = np.partition(np.abs(dct_mod[i, :]), -n_components)[-n_components]
    # Zero out smaller coefficients
    dct_mod[i, :] = np.where(np.abs(dct_mod[i, :]) >= threshold, dct_mod[i, :], 0)

# Perform the inverse DCT on the modified coefficients
idct_result = idct(dct_mod, norm='ortho', axis=1)
```

## Independent component Analysis

```python
from sklearn.decomposition import FastICA
import numpy as np
import matplotlib.pyplot as plt

# Normalize your data (necessary for ICA)
data_normalized = (df - df.mean()) / df.std()

# Choose a number of components - this should be as high as computationally feasible to ensure all significant components are captured
n_components = min(len(df), len(df.columns))  # or some other high number

# Fit ICA
ica = FastICA(n_components=n_components, random_state=0)
S_ = ica.fit_transform(data_normalized)
A_ = ica.mixing_

# Calculate the importance of each component. Here we use the sum of absolute values of the mixing matrix coefficients.
importances = np.sum(np.abs(A_), axis=0)
sorted_importances = np.sort(importances)[::-1]


# Plot the importances on a log-log scale
plt.plot(sorted_importances, marker='o')
plt.xlabel('Component index')
plt.ylabel('Importance')
plt.title('Importance of ICA Components')
plt.show()
```

```python
from sklearn.decomposition import FastICA

ica = FastICA(n_components=22, random_state=42)  # Set the number of components to reduce to
reduced_data = ica.fit_transform(df)
reduced_data.shape
```

```python
plt.figure(figsize=(8, 4))
plt.imshow(reduced_data)
plt.colorbar(label='Value')
plt.title('ICA')
plt.xlabel('Sample Index')
plt.ylabel(' Index')
plt.show()
```

```python
reconstructed_data = ica.inverse_transform(reduced_data)
plt.figure(figsize=(8, 4))
plt.imshow(reconstructed_data)
plt.colorbar(label='Value')
plt.title('ICA')
plt.xlabel('Sample Index')
plt.ylabel(' Index')
plt.show()
```
